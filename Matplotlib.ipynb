```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

#Colored Scatterplots

```python
iris = pd.read_csv('iris.csv')
iris.head()
```

```python
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'])
plt.xlabel('sepal lenghth')
plt.ylabel('petal lenghth')
```

```python
iris.sample(5)
```

```python
iris['Species'] = iris['Species'].replace({'Iris-setosa':0, 'Iris-virginica':1,'Iris-versicolor':2})
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'], c=iris['Species'])
plt.xlabel('sepal lenghth')
plt.ylabel('petal lenghth')
plt.colorbar()
```

#plot size

```python
plt.figure(figsize=(15, 7))
iris['Species'] = iris['Species'].replace({'Iris-setosa':0, 'Iris-virginica':1,'Iris-versicolo':2})
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'], c=iris['Species'])
plt.xlabel('sepal lenghth')
plt.ylabel('petal lenghth')
plt.colorbar()


```

#Annotations

```python
batters = pd.read_csv('batter (1).csv')
```

```python
batters.head()
```

```python
sample_df = batters.head(100).sample(25,random_state=5)
sample_df
```

```python
plt.scatter(sample_df['avg'], sample_df['strike_rate'])
```

```python
plt.figure(figsize=(18,10))
plt.scatter(sample_df['avg'],sample_df['strike_rate'],s=sample_df['runs'])

for i in range(sample_df.shape[0]):
  plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i])
```

#Horizontal and Vertical lines

```python
plt.figure(figsize=(18,10))
plt.scatter(sample_df['avg'],sample_df['strike_rate'],s=sample_df['runs'])
plt.axhline(130,color='red')
plt.axhline(140,color='green')
plt.axvline(30,color='red')
for i in range(sample_df.shape[0]):
  plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i])
```

```python
batters.head()
```

```python
#subplot
fig, axes = plt.subplots(nrows=2,ncols=1, sharex=True)
axes[0].scatter(batters['avg'], batters['strike_rate'],color='red')
axes[1].scatter(batters['avg'], batters['runs'])
axes[0].set_title('Avg vs Strike Rate')
axes[0].set_ylabel(' Strike Rate')

axes[1].set_title('Avg vs Runs')
axes[1].set_ylabel('Runs')
axes[1].set_xlabel('Avg')
```

```python
fig, axes = plt.subplots(nrows=2,ncols=2,figsize=(10,10))
axes[0,0].scatter(batters['avg'], batters['strike_rate'],color='red')
axes[0,1].scatter(batters['avg'], batters['runs'])
axes[1,0].hist(batters['avg'])
axes[1,1].hist(batters['runs'])

```

```python
fig = plt.figure()
ax1 = fig.add_subplot(2,1,1)
ax1.scatter(batters['avg'],batters['strike_rate'], color='red')

ax2 =fig.add_subplot(2,1,2)
ax2.hist(batters['runs'])
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(1,2,1)
ax1.scatter(batters['avg'],batters['strike_rate'], color='red')

ax2 =fig.add_subplot(1,2,2)
ax2.hist(batters['runs'])
```

```python
#from pandas dataframe
Batsman = pd.read_csv('sharma-kohli.csv')
Batsman.head()
```

```python
plt.plot(Batsman['index'],Batsman['V Kohli'])
```

```python
#ploting multiple plot
plt.plot(Batsman['index'],Batsman['V Kohli'])
plt.plot(Batsman['index'],Batsman['RG Sharma'])
```

```python
#labels title
plt.plot(Batsman['index'],Batsman['V Kohli'])
plt.plot(Batsman['index'],Batsman['RG Sharma'])

plt.title('Rohit Sharma vs Virat Kohli   (IPL Career Comparison )')
plt.xlabel('Season')
plt.ylabel('Runs scored')
```

```python
#colors
plt.plot(Batsman['index'],Batsman['V Kohli'], color='#F7927E')
plt.plot(Batsman['index'],Batsman['RG Sharma'], color='#542016')

plt.title('Rohit Sharma vs Virat Kohli   (IPL Career Comparison )')
plt.xlabel('Season')
plt.ylabel('Runs scored')
```

```python
#lines
plt.plot(Batsman['index'],Batsman['V Kohli'], color='#F7927E', linestyle='dashdot', linewidth=3)
plt.plot(Batsman['index'],Batsman['RG Sharma'], color='#542016', linestyle='dashdot',linewidth=2)

plt.title('Rohit Sharma vs Virat Kohli   (IPL Career Comparison )')
plt.xlabel('Season')
plt.ylabel('Runs scored')
```

```python
#marker
plt.plot(Batsman['index'],Batsman['V Kohli'], color='#F7927E', linestyle='dashdot', linewidth=3,marker='o',markersize=5)
plt.plot(Batsman['index'],Batsman['RG Sharma'], color='#542016', linestyle='dashdot',linewidth=2,marker='o',markersize=5)

plt.title('Rohit Sharma vs Virat Kohli   (IPL Career Comparison )')
plt.xlabel('Season')
plt.ylabel('Runs scored')
```

```python
#legend -> location
plt.plot(Batsman['index'],Batsman['V Kohli'], color='#F7927E', linestyle='dashdot', linewidth=3,marker='o',markersize=5,label='virat')
plt.plot(Batsman['index'],Batsman['RG Sharma'], color='#542016', linestyle='dashdot',linewidth=2,marker='o',markersize=5,label='rohit')

plt.title('Rohit Sharma vs Virat Kohli   (IPL Career Comparison )')
plt.xlabel('Season')
plt.ylabel('Runs scored')

plt.legend()
```

```python
#grid
plt.plot(Batsman['index'],Batsman['V Kohli'], color='#F7927E', linestyle='dashdot', linewidth=3,marker='o',markersize=5,label='virat')
plt.plot(Batsman['index'],Batsman['RG Sharma'], color='#542016', linestyle='dashdot',linewidth=2,marker='o',markersize=5,label='rohit')

plt.title('Rohit Sharma vs Virat Kohli   (IPL Career Comparison )')
plt.xlabel('Season')
plt.ylabel('Runs scored')

plt.legend()
plt.grid()
```

```python
#show
plt.plot(Batsman['index'],Batsman['V Kohli'], color='#F7927E', linestyle='dashdot', linewidth=3,marker='o',markersize=5,label='virat')
plt.plot(Batsman['index'],Batsman['RG Sharma'], color='#542016', linestyle='dashdot',linewidth=2,marker='o',markersize=5,label='rohit')

plt.title('Rohit Sharma vs Virat Kohli   (IPL Career Comparison )')
plt.xlabel('Season')
plt.ylabel('Runs scored')

plt.legend()
plt.grid()
plt.show()
```

```python
#color and label

df= pd.read_csv('batsman_season_record.csv')
df
```

```python
plt.bar(df['batsman'],df['2015'])
```

```python
plt.bar(np.arange(df.shape[0]) - 0.5,df['2015'])
```

```python
plt.bar(np.arange(df.shape[0]) - 0.2,df['2015'],width=0.2,color='#EB836E')
plt.bar(np.arange(df.shape[0]),df['2016'],width=0.2,color='#544C4B')
plt.bar(np.arange(df.shape[0]) + 0.2,df['2017'],width=0.2,color='#8EC5ED')

plt.xticks(np.arange(df.shape[0]), df['batsman'])

plt.show()
```

```python
children = [10,20,40,10,30]
colors = ['red red red red red red','blue blue blue blue','green green green green green','yellow yellow yellow yellow ','pink pinkpinkpink']

plt.bar(colors,children,color='#2F4354')
plt.xticks(rotation='vertical')
plt.show()
```

```python
# Stacked Bar chart
plt.bar(df['batsman'],df['2017'],label='2017')
plt.bar(df['batsman'],df['2016'],bottom=df['2017'],label='2016')
plt.bar(df['batsman'],df['2015'],bottom=(df['2016'] + df['2017']),label='2015')

plt.legend()
plt.show()
```
